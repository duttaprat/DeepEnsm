```python
import torch
import numpy as np
import pandas as pd
```

```python
search_with = pd.read_csv("preprocessed_BCLL.txt", sep="\t", header = None)
data1 = pd.read_csv("labeled_file.txt", sep="\t", header = None)
data2 = pd.read_csv("Labelled_prev_unlabelBCLL_CNN_M1.txt", sep=" ", header = None)
```

```python
print(search_with.head())
print(data1.head())
print(data2.head())
```

```python
search_from = pd.concat([data1, data2])
print(search_with.shape, search_from.shape)
```

```python
result = np.empty((search_from.shape[0], 1), dtype=np.long)
print(result.shape)
```

```python
c = 0
dict_sw = []
dict_sf = {}
for i in range(search_with.shape[0]):
    val = search_with.iloc[i, 0]
    val2 = search_from.iloc[i, 0]
    label = search_from.iloc[i, 23]
    if val not in dict_sw:
        dict_sw.append(val)
    else:
        print("COPY DETECTED. EXITING")
        break
    if val2 not in dict_sf:
        dict_sf[val2] = label
    else:
        print("COPY DETECTED. EXITING")
        break
```

```python
print(len(dict_sw), len(dict_sf))
```

```python
print(dict_sw)
```

```python
c = 0
for val in dict_sw:
    result[c] = dict_sf[val]
    c += 1
    if not val in dict_sf:
        print("NO")
        break
    print("FOUND! {} and {}".format(val, val in dict_sf))
```
