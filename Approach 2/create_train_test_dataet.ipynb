```python
from itertools import chain
with open("disconnected_walk.txt") as f:
    contents = f.read()
    contents = contents.replace('\n','')
    count = contents.split('set')
    
labels= []
label_no = []
for k in range(1, len(count)):
    temp = count[k].split(',')
    temp_list = []
    for i in temp:
        i=i.strip('])([')
        if (len(i)>0):
            temp_list.append(int(i))
    
    if (len(temp_list)>3):
        label_no.append(len(temp_list))
        labels.append(temp_list)

print label_no 
print "Number of sets", len(label_no)
print labels
print set(labels[0]).intersection(set(labels[1]))
```

```python
read_file = open('preprocessed_BCLL.txt', 'r')
file_content = read_file.read()
all_data = file_content.splitlines()
read_file.close()
```

```python
labelled_file  = open('labeled_file.txt', 'w+')
for k in range(len(labels)):
    for j in labels[k]:
        #print all_data[j], k
        labelled_file.write("%s" % all_data[j])
        labelled_file.write("%s\n" % k)
labelled_file.close()
```
